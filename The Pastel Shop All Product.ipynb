```python
import scrapy

class ThePastelShop(scrapy.Spider):
    name = 'thepastelshop'
    allowed_domains = ['thepastelsshop.com']  # Correct domain (no path needed)
    start_urls = ['https://www.thepastelsshop.com/collections/shop-by-category']  # Correct start URL

    def parse(self, response):
        # Corrected the XPath to be relative to the container
        product_container = response.xpath('//div[@class="card-wrapper"]')
        
        for container in product_container:
            # Use relative XPath for product name, and get() to fetch the first match
            name = container.xpath('.//span[@class="card-information__text h5"]/text()').get()
            
            # Use relative XPath for original and sale prices
            original = container.xpath('.//s[@class="price-item price-item--regular"]/span/text()').get('').strip()
            sale = container.xpath('.//span[@class="price-item price-item--sale"]/span/text()').get('').strip()
            
            # Yield the data
            yield {
                'name': name,
                'original': original,
                'sale': sale,
            }
```

```python
from scrapy.crawler import CrawlerProcess
process = CrawlerProcess(settings = {'FEEDS': {'tpsallproduct.csv':{'format':'csv'}}})
process.crawl(ThePastelShop)
process.start()
```
